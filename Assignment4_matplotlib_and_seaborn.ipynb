## Matplotlib

```python
# Task 1: Basic Matplotlib (3 plots)
# 1. Bar chart of gym membership types
# Use: gym['abonoment_type'].value_counts()

# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])

# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])
```

#### Bar chart

```python
# 1. Bar chart of gym membership types
# Use: gym['abonoment_type'].value_counts()

abonoment_count = gym['abonoment_type'].value_counts()
print(abonoment_count)

plt.bar(abonoment_count.index, abonoment_count.values)
plt.xlabel('Membership Type')
plt.ylabel('Count')
plt.title('Count of Membership Types')
plt.show()
```

#### Histogram

```python
# 2. Histogram of time in gym
# Use: plt.hist(gym['avg_time_in_gym'])

plt.hist(gym['avg_time_in_gym'],bins=20)
plt.title("Histogram of average time in gym")
plt.xlabel("Time in gym")
plt.ylabel("Frequency")
plt.show()
```

#### Scatter plot

```python
# 3. Scatter plot: Age vs visits per week
# Use: plt.scatter(gym['Age'], gym['visit_per_week'])

plt.scatter(gym['Age'],gym['visit_per_week'])
plt.grid(True)
plt.title("Age vs visits per week")
plt.xlabel("Age")
plt.ylabel("Visits per week")
plt.show()
```

#### Pie chart

```python
abm_type = gym['abonoment_type'].value_counts()
print(abm_type)

plt.pie(abm_type.values, labels=abm_type.index, autopct='%1.1f%%')
plt.title('Abonoment type')
plt.show()
```

## Seaborn

```python
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')

# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')

# 3. Bar plot: Average visits by membership type
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')
```

#### Box plot

```python
# 1. Box plot: Gender vs Age
# Use: sns.boxplot(data=gym, x='gender', y='Age')

sns.boxplot(data=gym, x='gender',y='Age')
plt.grid(True,which='both',linewidth=0.7,alpha=0.7,linestyle='--')
plt.title("Gender vs Age")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```

#### Count plot

```python
# 2. Count plot: Days per week
# Use: sns.countplot(data=gym, x='days_per_week')

sns.countplot(data=gym,x='days_per_week')
plt.grid(True,which='both',linewidth=0.7,alpha=0.7,linestyle='--')
plt.title("Days per week")
plt.xlabel("Days")
plt.ylabel("Count")
plt.show()
```

#### Bar plot

```python
# 3. Bar plot: Average visits by membership type
# Use: sns.barplot(data=gym, x='abonoment_type', y='visit_per_week')

sns.barplot(data=gym,x='abonoment_type',y='visit_per_week')
plt.grid(True,which='both',linewidth=0.7,alpha=0.7)
plt.title("Average visits by membership type")
plt.xlabel("Membership type")
plt.ylabel("Average visits")
plt.show()
```

#### Kernel Density Estimate plot

```python
sns.kdeplot(data=gym,x='Age',hue='gender',fill=True)
plt.grid(True,which='both',linewidth=0.7,alpha=0.7)
plt.title("Kernel Density Estimate plot")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

#### Heatmap

```python
gym_numerical = gym[['Age','visit_per_week','avg_time_in_gym']]
correlation = gym_numerical.corr()
print(round(correlation,2))

sns.heatmap(correlation,annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

#### Pair plot

```python
sns.pairplot(gym_numerical)
plt.show()
```
